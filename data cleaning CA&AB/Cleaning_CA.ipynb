```python
import pandas as pd
import json

def flatten_dict(d, parent_key='', sep='_'):
    """
    Recursively flattens a nested dictionary.
    
    Parameters:
    - d: dict, the dictionary to flatten.
    - parent_key: str, the base key string for nested keys.
    - sep: str, the separator between parent and child keys.
    
    Returns:
    - A flat dictionary where nested keys are concatenated into a single key.
    """
    items = {}
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.update(flatten_dict(v, new_key, sep=sep))
        else:
            items[new_key] = v
    return items

def normalize_json_data(data, record_path='items'):
    """
    Normalizes JSON data by flattening nested dictionaries for each record.
    
    Parameters:
    - data: dict, the loaded JSON data.
    - record_path: str, key where the list of records is stored (e.g., 'items').
    
    Returns:
    - A pandas DataFrame with flattened columns.
    """
    if record_path not in data:
        raise ValueError(f"Record path '{record_path}' not found in data.")
    
    # Extract and flatten each record under the specified record path.
    records = data[record_path]
    flattened_records = [flatten_dict(record) for record in records]
    df = pd.DataFrame(flattened_records)
    
    top_level = {k: v for k, v in data.items() if k != record_path}
    for key, value in top_level.items():
        df[key] = value
        
    return df


with open('CA_category_id.json') as f:
    data_json = json.load(f)



# Normalize and flatten the JSON data automatically.
df_clean = normalize_json_data(data_json)
print("Normalized DataFrame:")
display(df_clean)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

# Check for missing values
print("Missing values in each column:")
print(df_clean.isnull().sum())
```

```python
# If needed, drop rows with missing values (or alternatively, you can impute them)
df_clean.dropna(inplace=True)

# Remove duplicate rows (if any)
df_clean.drop_duplicates(inplace=True)
```

```python
# Standardize column names (convert to lower case and replace spaces with underscores)
df_clean.columns = df_clean.columns.str.lower().str.replace(' ', '_')
```

```python
# Convert 'id' to numeric if it makes sense (here, id is a string from the API but can be converted)
# We'll use errors='ignore' in case conversion isn't needed.
df_clean['id'] = pd.to_numeric(df_clean['id'])

print("\nCleaned DataFrame:")
display(df_clean)
```

```python
df_clean.rename(columns={
    'snippet_channelid': 'channel_id',
    'snippet_title': 'category_title',
    'snippet_assignable': 'assignable'
}, inplace=True)

# Display the updated DataFrame columns
print("Updated DataFrame columns:")
print(df_clean.columns)

# Improve label clarity and chart layout
plt.title('Count of Video Categories by Title', fontsize=14)
plt.xlabel('Video Category Title', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
